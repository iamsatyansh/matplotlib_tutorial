Loading neccessary libraries

```python
import matplotlib.pyplot as plt  # Matplotlib for plotting
import numpy as np  # NumPy for numerical computing
import pandas as pd  # Pandas for data manipulation
```

Basic graph


```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

# Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

Bar Chart

```python
# Define labels and corresponding values for the bar chart
labels = ['A', 'B', 'C']  # Labels for the bars
values = [1, 4, 2]  # Corresponding values for each label

# Resize the figure (dpi specifies pixels per inch)
plt.figure(figsize=(5, 3), dpi=100)

# Create the bar chart
bars = plt.bar(labels, values)

# Define patterns for each bar in the chart
patterns = ['/', 'O', '*']

# Iterate over each bar and apply a pattern
for bar in bars:
    bar.set_hatch(patterns.pop(0))  # Pop the first pattern from the list and apply it to the current bar

# Save the figure as an image file with high resolution
plt.savefig('barchart.png', dpi=300)

# Display the plot
plt.show()
```

Line Graph

```python
# Load the gas prices data from a CSV file into a pandas DataFrame
gas = pd.read_csv('gas_prices.csv')

# Resize the figure
plt.figure(figsize=(8, 5))

# Set the title of the plot with specified font parameters
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight': 'bold', 'fontsize': 18})

# Plot gas prices for different countries over time
plt.plot(gas.Year, gas.USA, 'b.-', label='United States')  # Plot gas prices for the USA
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')  # Plot gas prices for Canada
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')  # Plot gas prices for South Korea
plt.plot(gas.Year, gas.Australia, 'y.-', label='Australia')  # Plot gas prices for Australia

# Set custom tick marks for the x-axis (showing every 3rd year)
plt.xticks(gas.Year[::3].tolist() + [2011])

# Set labels for the x-axis and y-axis
plt.xlabel('Year')
plt.ylabel('US Dollars')

# Add a legend to distinguish between plotted lines
plt.legend()

# Save the figure as an image file with high resolution
plt.savefig('Gas_price_figure.png', dpi=300)

# Display the plot
plt.show()
```

Load Fifa Data

```python
# Load FIFA data from a CSV file into a pandas DataFrame
fifa = pd.read_csv('fifa_data.csv')

# Display the first three rows of the DataFrame
print(fifa.head(3))

plt.show()
```

Histogram


```python
# Define the bins for the histogram
bins = [40, 50, 60, 70, 80, 90, 100]

# Create a new figure for the histogram with specified size
plt.figure(figsize=(8, 5))

# Plot the histogram of player overall skill levels
plt.hist(fifa.Overall, bins=bins, color='#abcdef')

# Set custom tick marks for the x-axis
plt.xticks(bins)

# Set labels for the x-axis and y-axis
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')

# Set the title of the histogram
plt.title('Distribution of Player Skills in FIFA 2018')

# Save the figure as an image file with high resolution
plt.savefig('histogram.png', dpi=300)

# Display the histogram plot
plt.show()
```

Pie chart #1


```python
# Count the number of players with left and right preferred foot
left = len(fifa.loc[fifa['Preferred Foot'] == 'Left'])
right = len(fifa.loc[fifa['Preferred Foot'] == 'Right'])

# Create a new figure for the pie chart with specified size
plt.figure(figsize=(8, 5))

# Define labels and colors for the pie chart
labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

# Plot the pie chart showing the distribution of preferred foot
plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f %%')

# Set the title of the pie chart
plt.title('Foot Preference of FIFA Players')

# Display the pie chart
plt.show()
```

Pie Chart #2

```python
# Create a new figure for the pie chart with specified size and DPI
plt.figure(figsize=(8, 5), dpi=100)

# Set the style of the plot to 'ggplot'
plt.style.use('ggplot')

# Clean the 'Weight' column by removing 'lbs' and converting to integer
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

# Count the number of players in different weight categories
light = len(fifa.loc[fifa.Weight < 125])
light_medium = len(fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)])
medium = len(fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)])
medium_heavy = len(fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)])
heavy = len(fifa[fifa.Weight >= 200])

# Define weights and labels for the pie chart
weights = [light, light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']

# Define explode values to separate slices
explode = (0.4, 0.2, 0, 0, 0.4)

# Set the title of the pie chart
plt.title('Weight of Professional Soccer Players (lbs)')

# Plot the pie chart with specified labels, explode, and percentage formatting
plt.pie(weights, labels=label, explode=explode, pctdistance=0.8, autopct='%.2f %%')

# Display the pie chart
plt.show()
```
